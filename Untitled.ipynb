```python
import torch

from torch import nn
import matplotlib.pyplot as plt
```

```python
def return_linear():
    linear = nn.Linear(10, 10, bias=False)
    torch.nn.init.constant_(linear.weight, 0.5)
    return linear
```

```python
lr = 0.1

net = nn.Sequential(return_linear(), return_linear())
opt = torch.optim.SGD(lr=lr, params=net.parameters())

one = return_linear()
opt1 = torch.optim.SGD(lr=lr, params=one.parameters())
two = return_linear()
opt2 = torch.optim.SGD(lr=lr, params=two.parameters())
```

```python
losses_tot = []
losses_sin = []
for _ in tqdm(range(1000)):
    i = torch.rand(10)
    o_tot = net(i)
    loss_tot = o_tot.square().mean()
    losses_tot.append(loss_tot.detach())
    
    opt.zero_grad()
    loss_tot.backward()
    opt.step()
    
    o_sin = two(one(i))
    loss_sin = o_sin.square().mean()
    losses_sin.append(loss_sin.detach())
    
    opt1.zero_grad()
    opt2.zero_grad()
    loss_sin.backward()
    opt1.step()
    opt2.step()
```

```python
plt.plot(losses_tot)
plt.plot(losses_sin)
plt.yscale("log")
```
